```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = np.arange(10)
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(),'k--')
ax1.hist(np.random.randn(100),bins=20,color="k",alpha=0.3)
ax2.scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))
```

```python
fig,axes = plt.subplots(2,3,sharex=True,sharey=False)
for i in range(2):
    for j in range(3):
        axes[i,j].hist(np.random.randn(500),bins=20)
plt.subplots_adjust(wspace=0,hspace=0)
```

```python
plt.plot(np.random.randn(30),color = "r",linestyle = "dashed",marker="o")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(),"k--",label="one")
ax.plot(np.random.randn(1000).cumsum(),"r-",label="two")
ax.legend(loc="best")#plt.legend()ax.legend()
ax.text(100,10,"hello world",family="monospace",fontsize=10)
plt.savefig("test3.pdf",dpi=500)
```
